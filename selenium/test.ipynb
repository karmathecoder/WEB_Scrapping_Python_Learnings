## Importing Selenium WebDriver

```python
from selenium import webdriver
```

##### Initializing Session

```python
driver = webdriver.Chrome(executable_path='webdriver/chrome/chromedriver.exe')
driver
```

#### Interacting 

```python
driver.get('https://mirrorapp.mirroraccone.workers.dev/0:/')
```

```python
driver.title
```

##### Going Back

```python
driver.back()
```

##### Going Forward

```python
driver.forward()
```

##### Getting ScreenShots

```python
driver.save_screenshot('screen.png')
```

### Getting Page Source

```python
driver.page_source
```

### Checking or searching For a button

```python
from selenium.webdriver.common.by import By
```

```python
course_button = driver.find_element(By.LINK_TEXT,'Courses')
course_button
```

```python
course_button.click()
```

```python
course_button = driver.find_element(By.PARTIAL_LINK_TEXT,"Cour")
```

```python
course_button.click()
```
