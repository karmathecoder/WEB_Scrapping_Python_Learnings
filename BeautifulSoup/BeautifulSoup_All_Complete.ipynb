# Loading Data

```python
import requests
```

```python
response = requests.get('https://www.w3schools.com/html/html_basic.asp')
```

```python
data = response.content
```

# Parsing string data to BeautifulSoup

```python
from bs4 import BeautifulSoup
```

```python
parse_data = BeautifulSoup(data, 'html.parser')
parse_data
```

```python
type(parse_data)
```

```python
print(parse_data.prettify())
```

## Acessing using tag name (data.tag_name)

```python
parse_data.title
```

```python
parse_data.head
```

#### If multiple Tag then we get the very first tag

```python
parse_data.h1
```

# Acessing using data.tagname.name or .string or .attributes

```python
print(parse_data.title)
print(parse_data.title.name)
print(parse_data.title.string)
```

```python
print(parse_data.div.attrs)
```

```python
parse_data.div['id']
```

```python
parse_data.div.get('id')
```

## getting just text

```python
print(parse_data.get_text())
```

## Find Methods

```python
parse_data.p
```

```python
parse_data.find('p')
```

##### Ocuernce of all P tag returned as list

```python
parse_data.find_all('p')
```

```python
parse_data.find_all(True)
```
